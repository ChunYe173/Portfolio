```python
from transformers import pipeline
```

```python
# Initializing pretrained model (ESGify)

esgify_classifier = pipeline("text-classification", model="ai-lab/ESGify")
```

```python
# Writing a function that take a text as input and output a ESG topic based on the context

def esg_classification(text):
  predictions = esgify_classifier(text)
  esg_label = predictions[0]['label']
  return esg_label
```

```python
# Example text took from Amazon's sustainability report 2022

text1 = "Decrease in absolute carbon emissions, as year-on-year net sales grew 9%"
predict1 = esg_classification(text1)
print('Prediction Label:', predict1)
```

```python
dataset = load_dataset('Jean-Baptiste/financial_news_sentiment_mixte_with_phrasebank_75')
```

```python
import pandas as pd

df = pd.DataFrame(dataset['train'])
df.head()
```

```python
# Take 10 rows for sample and test the robustness of the classification model

df_subset = df.loc[0:10]
df_subset['ESG Topics'] = df_subset['summary_detail_with_title'].apply(esg_classification)
```
